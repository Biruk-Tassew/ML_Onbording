```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./imdb_top_1000.csv")

columns = df.columns.tolist()

average_rating = df["IMDB_Rating"].mean()

unique_genre = df['Genre'].unique()
```

```python
unique_df = pd.DataFrame(unique_genre, columns=["Unique Genre"])

# Save the unique values to a CSV file
output_file = './genre_file.csv'
unique_df.to_csv(output_file, index=False)
```

```python
fig = plt.figure(figsize=(18, 6))

plt.hist(df['Genre'], bins=10)
plt.xlabel('Unique Genre')
plt.ylabel('Frequency')
plt.title('Histogram of Genre')
plt.xticks(rotation=90, ha='center')
plt.show()
```
